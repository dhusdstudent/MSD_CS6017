```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
def computeMean(data):
    mean = 0

    for x in data:
        mean += x

    return mean / len(data)

def standardDeviation(data):
    mean = computeMean(data)
    addup = 0

    for x in data:
        addup += (x - mean) ** 2

    variance = addup / len(data)
    return np.sqrt(variance)
```

```python
from scipy.stats import norm
import numpy as np

sample = norm.rvs(loc=50, scale=10, size=1000)

myMean = computeMean(sample)
myStand = standardDeviation(sample)

numpyMean = np.mean(sample)
numpyStand = np.std(sample)

print("Regular mean: ", myMean)
print("Standard deviation: ", myStand)

print("Numpy mean: ", numpyMean)
print("Numpy standard deviation: ", numpyStand)
```

```python
import matplotlib.pyplot as plt

plt.hist(sample, bins=50)

plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

Null Hypothesis: H0 --> Pa = Pb (representing the idea that both perform the same)
Data-driven Hypothesis H1 -->Pb > Pa (based on the actual clicks)
    Pa = .5 and Pb = .55 (halfway between = .525)

```python
import numpy as np
from scipy.stats import norm

a_clicks = 500
b_clicks = 550

a_views = 1000
b_views = 1000

pA = a_clicks / a_views
pB = b_clicks / b_views

prop = (a_clicks + b_clicks) / (a_views + b_views)

variation = np.sqrt(prop * (1 - prop) * ((1/a_views) + (1/b_views)))

ZZ = (pB - pA) / variation
sigVal = 1 - norm.cdf(ZZ)

print("Icon A:", pA)
print("Icon B:", pB)
print("Proportion:", prop)
print("Variation:", variation)
print("Z-Value:", ZZ)
print("Significance:", sigVal)
```
