```python
## --------------------------Year-Long Data
import pandas as pd
import matplotlib.pyplot as plt

yearFile = pd.read_csv("Year2013.csv", header=2)
yearFile.columns = ["Date","BR","HA","HC", "HW","L4","LN","NP","O2","RS","T3","VL"]
yearFile["Date"] = pd.to_datetime(yearFile["Date"])
##print(yearFile.head())

plt.figure(figsize=(15,5))
plt.plot(yearFile["Date"], yearFile["L4"])

plt.title("Station L4 Levels in 2013")
plt.xlabel("Date")
plt.ylabel("PM2.5 (UG/M3)")
plt.show()
```

```python
## --------------------------Monthly Mean Data
import pandas as pd
import matplotlib.pyplot as plt

yearFile = pd.read_csv("Year2013.csv", header=2)
yearFile.columns = ["Date","BR","HA","HC", "HW","L4","LN","NP","O2","RS","T3","VL"]
yearFile["Date"] = pd.to_datetime(yearFile["Date"])

yearFile["Month"] = yearFile["Date"].dt.month

monthMean = yearFile.groupby("Month")["L4"].mean()

monthMean.plot(kind="bar", figsize=(15,5), color="blue")

plt.title("Monthly PM2.5 Lvls at L4 Station")
plt.xlabel("Month")
plt.ylabel("Average")
plt.show()

##From the new, zoomed in bar graph (below) it's much easier to visually parse the data and notice trends based on the month over month results.
```

```python
## --------------------------Hourly Mean Data
import pandas as pd
import matplotlib.pyplot as plt

yearFile = pd.read_csv("Year2013.csv", header=2)
yearFile.columns = ["Date","BR","HA","HC", "HW","L4","LN","NP","O2","RS","T3","VL"]
yearFile["Date"] = pd.to_datetime(yearFile["Date"])

yearFile["Hour"] = yearFile["Date"].dt.hour

hourMean = yearFile.groupby("Hour")["L4"].mean()

hourMean.plot(kind="bar", figsize=(15,5), color="green")

plt.title("Hourly PM2.5 Lvls at L4 Station")
plt.xlabel("Hour")
plt.ylabel("Average")
plt.show()

## In addition to also being broken down more thoroughly, this sort of view helps us to establish the general patterns across a smaller window of time, basing the results on a multitude of hourly data patterns.
```

```python
## --------------------------Monthly Average
import pandas as pd
import matplotlib.pyplot as plt

yearFile = pd.read_csv("Year2013.csv", header=2)
yearFile.columns = ["Date","BR","HA","HC", "HW","L4","LN","NP","O2","RS","T3","VL"]
yearFile["Date"] = pd.to_datetime(yearFile["Date"])

yearFile["Month"] = yearFile["Date"].dt.month

yearFile.boxplot(column="L4", by="Month", figsize=(15,5))

plt.title("Monthly PM2.5 Distribution")
plt.suptitle("")
plt.xlabel("Month")
plt.ylabel("PM2.5")
plt.show()

##While I personally find this a little more noisy visually, it certainly packs more information into a singular instance of a graph without being too intimidating (as per the first graph). It gives us a more granular month by month breakdown
```

```python
## --------------------------Exercise 4
import pandas as pd
import matplotlib.pyplot as plt

yearFile = pd.read_csv("Year2013.csv", header=2)
yearFile.columns = ["Date","BR","HA","HC", "HW","L4","LN","NP","O2","RS","T3","VL"]
yearFile["Date"] = pd.to_datetime(yearFile["Date"])

yearFile["Hour"] = yearFile["Date"].dt.hour

yearFile.boxplot(column="L4", by="Hour", figsize=(15,5))

plt.title("Hourly PM2.5 Distribution (L4)")
plt.xlabel("Hour")
plt.ylabel("PM2.5")
plt.show()

##In my personal opinion, this sort of graph is less suited for such a precise series of measurements. The circles often crowd together to the point that they're difficult to interpret, leading me to speculate that this graph format would be preferable for data such as the month-to-month interpretation, while something more static like the bar graph would be preferable for smaller increments. 
```
